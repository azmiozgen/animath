```python
import gizeh
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 10)
```

```python
# initialize surface
surface = gizeh.Surface(width=320, height=260) # in pixels
```

```python
Pi = 3.14159265
circle = gizeh.circle(r=30, xy=(50,50), fill=(0,0,0))
rectangle = gizeh.rectangle(lx=60.3, ly=45, xy=(60,70), fill=(0,1,0), angle=Pi/8)
square = gizeh.square(l=20, stroke=(0,0,0), stroke_width= 1.5)
arc = gizeh.arc(r=20, a1=Pi/4, a2=3*Pi/4, fill=(0,1,0))
text = gizeh.text("Hello world", fontfamily="Impact",  fontsize=40,
                  fill=(0,0,0), xy=(200,100), angle=Pi/12)
polygon = gizeh.regular_polygon(r=40, n=5, angle=np.pi/4, xy=[180,100], fill=(1,0,1))
line = gizeh.polyline(points=[(0,200), (20,230), (40,240), (0,210)], stroke_width=3,
                     stroke=(0.5,0.73,0.745), fill=(1,0,1))
```

```python
circle.draw(surface)
rectangle.draw(surface)
square.draw(surface)
arc.draw(surface)
text.draw(surface)
polygon.draw(surface)
line.draw(surface)

# Now export the surface
image = surface.get_npimage() # returns a (width x height x 3) numpy array
```
